MERGE IN DATAFRAME

```python
#import libraries

import pandas as pd

#create DF's
df1=pd.DataFrame({"A":[80,85,8],"B":[1,2,3]})
df2=pd.DataFrame({"A":[80,85,8],"C":[1,2,3]})

#perform merge
pd.merge(df1,df2)
```

```python
#using how
#create DF's
df1=pd.DataFrame({"A":[1,2,3,4],"B":[11,21,31,41]})
df2=pd.DataFrame({"A":[1,2,3,5],"C":[10,20,30,40]})

#perform merge
pd.merge(df1,df2,on="A",how="inner")
```

```python
df1=pd.DataFrame({"A":[1,2,3,4],"B":[11,21,31,41]})
df2=pd.DataFrame({"A":[1,2,3,5],"C":[10,20,30,40]})

#perform merge
pd.merge(df1,df2,on="A",how="outer")
```

```python
df1=pd.DataFrame({"A":[1,2,3,4],"B":[11,21,31,41]})
df2=pd.DataFrame({"A":[1,2,3,5],"C":[10,20,30,40]})

#perform merge
pd.merge(df1,df2,on="A",how="right")
```

```python
df1=pd.DataFrame({"A":[1,2,3,4],"B":[11,21,31,41]})
df2=pd.DataFrame({"A":[1,2,3,5],"C":[10,20,30,40]})

#perform merge
pd.merge(df1,df2,on="A",how="left")
```
