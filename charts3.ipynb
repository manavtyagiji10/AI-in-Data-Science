```python
from matplotlib.lines import lineStyles
Sales={"Months":["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"],
"Sales_unit":[20,21,22,23,24,25,26,35,34,34,36,38],
"Revenue_in_lakh":[3,5,6,7,8,9,12,15,16,18,22,25]}

import matplotlib.pyplot as plt
import pandas as pd
plt.plot(Sales["Months"],Sales["Revenue_in_lakh"],marker="o")
plt.xlabel("Months",color="green",fontsize="18")
plt.ylabel("Sales_unit",color="green",fontsize="18")
plt.legend(["Revenue_in_lakh"])
plt.grid(True)
plt.show()
```

```python
plt.bar(Sales['Months'],Sales['Revenue_in_lakh'],color="red")
plt.show()
```

```python
plt.hist(Sales['Revenue_in_lakh'],color="green")
plt.xlabel("Revenue_in_lakh")
plt.ylabel("Frequency")
plt.legend(["Revenue_in_lakh"])
plt.grid(True)
plt.show()
```

```python
q1=sum(Sales["Revenue_in_lakh"][0:3])
q2=sum(Sales["Revenue_in_lakh"][3:6])
q3=sum(Sales["Revenue_in_lakh"][6:9])
q4=sum(Sales["Revenue_in_lakh"][9:12])
plt.pie([q1,q2,q3,q4],labels=["Q1","Q2","Q3","Q4"],autopct="%1.2f%%",explode=(0,0,0.1,0))
plt.title("Quarterly Revenue")
plt.legend(["Q1","Q2","Q3","Q4"])
plt.show()
```

```python
plt.scatter(Sales["Months"],Sales["Sales_unit"],color="red",marker="s")
plt.xlabel("Months")
plt.ylabel("Sales_unit")
plt.legend(["Sales_unit"])
plt.grid(True)
plt.show()
```
