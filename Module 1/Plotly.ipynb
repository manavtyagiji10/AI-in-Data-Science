Data Vizulation with- Plotly

```python
#libraries
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
India_data=px.data.gapminder().query("country=='India'")
India_data
```

```python
#bar_graph
graph=px.bar(India_data,x='year',y='pop',title="Population_of_India",labels={'pop':"Population"},color='pop',hover_data=['lifeExp','gdpPercap'],text='pop',template='plotly_dark')
graph.update_traces(texttemplate='%{text:.2s}',textposition='inside')
graph.update_layout(yaxis_type='log',xaxis_tickangle= -45)
graph.show()
```

```python
graph=px.line(India_data,x='year',y='pop',title="Population_India_Overtheyear",labels={'pop':"Population"},color='pop',hover_data=['lifeExp','gdpPercap'],text='pop')
graph.update_traces(texttemplate='%{text:.2s}')
graph.show()
```

```python
graph=px.scatter(India_data,x='gdpPercap',y='lifeExp',size='pop',title="PerCap vs LifeExp",labels={'pop':"Population"},color='pop',hover_data=['lifeExp','gdpPercap'],text='pop',template='plotly_dark')
graph.update_traces(texttemplate='%{text:.2s}')
graph.show()
```

```python
#pie chart
px.pie(India_data,values='pop',names='year',title='Population_of_India',hole=0.5)
```

```python
#heat map
graph=px.density_heatmap(India_data,x='year',y='pop',title='Heatmap',template='plotly_dark')
graph.show()
```

```python
#hist graph
graph=px.histogram(India_data,x='gdpPercap',y='lifeExp',title='Histogram',template='plotly_dark')
graph.show()
```

```python
#boxplot
graph=px.box(India_data,y='pop',title='Boxplot',points='all',color_discrete_sequence=['blue'])
graph.update_traces(boxmean=False,jitter=0.1,marker=dict(size=7),opacity=0.5)
graph.show()
```

```python
#violen graph
graph=px.violin(India_data,y='gdpPercap',title='Violin',template='plotly_dark')
graph.show()
```
